```python
# CARREGAR O ARQUIVO
import pandas as pd 

df = pd.read_excel("dados_ausentes.xlsx")
print("\nPlanilha Dados Ausentes")
print(df)
```

```python
# Verificar se existem dados faltando
# isnull()
print(df.isnull())
```

```python
# Contar o numero de ausência por coluna 

print(df.isnull().sum())
```

```python
# Excluir a linha dos dados ausentes

df_sem_vazio = df.dropna()
print(df_sem_vazio)
```

```python
# Substituir dados ausentes 
# Cidade Vazia - desconhecido
# fillna - Preenche ausencias

df["Cidade"] = df["Cidade"].fillna("Desconhecido")
print(df)
```

```python
# Substituir dados ausentes
# Substituir "Salário" por 0

df["Salário"] = df["Salário"].fillna(0)
print(df)
```

```python
# Substituir dados ausentes
# Substituir "Idade" por média

df["Idade"] = df["Idade"].fillna(df["Idade"].mean())
print(df)
```

```python
# Corrijir os erros de formata;áo ( minusculos e maiusculos )

df["Nome"] = df["Nome"].str.capitalize()
df["Cidade"] = df["Cidade"].str.title()

print(df)
```

```python
# Padronização de nomenclatura 
# replace - substitui o valor 
df["Cidade"] = df["Cidade"].str.upper().replace({"são paulo": "SP",
                                                 "sao paulo": "SP",
                                                 "São Paulo": "SP",
                                                 "Sao Paulo": "SP",
                                                 "SÃO PAULO": "SP",
                                                 "SÃO PAULO ": "SP",
                                                 "SAO PAULO": "SP",
                                                 "RIO DE JANEIRO": "RJ",
                                                 "rio de janeiro": "RJ",})
print(df)
```

```python
# Identificar o tipo de dado em uma coluna 
# Tipo String

coluna_string = df.select_dtypes(include=["string"]).columns.to_list()
print(coluna_string)
```

```python
# Trocar o tipo de dado de idade para INT
df["Idade"] = df["Idade"].astype(int)
print(df)
```

```python
# Pegar o valor de cima e substituir a ausencia
df["Idade"] = df["Idade"].ffill()

# Debaixo
df["Idade"] = df["Idade"].bfill()
```

```python
# Criar uma nova coluna: Faixa Etária
df["Faixa Etária"] = pd.cut(df["Idade"], bins=[0, 18, 40, 60, 100], labels=["Jovem", "Adulto", "Meia-Idade", "Idoso"])
print(df)
```

```python
# Remover duplicatas 
df_sem_duplicata = df.drop_duplicates(["Nome"])
print(df_sem_duplicata)
```

```python
# Montar categoria de salário
df["Faixa Salarial"] = df["Salário"].map(lambda x: "Baixo" if x <= 1000 else ("Médio" if x <= 3500 else "Alto"))
print(df)
```

```python
df.to_excel("dados_tratados.xlsx", index=False)
```
